# 导包

```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = cv2.imread('images/highway.jpg', -1)
```

# matlpotlib 图像显示

```python
def show(img, title='Image'):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # 不显示坐标轴
    plt.title(title) # 图像题目
    plt.show()
```

```python
show(fig, 'Car in highway')
```
