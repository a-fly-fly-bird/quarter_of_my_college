```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img, title='Image'):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # 不显示坐标轴
    plt.title(title) # 图像题目
    plt.show()
```

```python
img = np.zeros((512, 512, 3), dtype = np.uint8)
show(img)
```

```python
fig = cv2.imread('images/highway.jpg', -1)
```

```python
fig.shape
```

```python
fig_cp = fig.copy()
for i in range(fig_cp.shape[0]):
    for j in range(fig_cp.shape[1]):
      for k in range(fig_cp.shape[2]):
        fig_cp[i][j][k] = 255 - fig_cp[i][j][k]
```

```python
show(fig_cp)
```

```python
fig_cp = fig.copy()
fig_cp[:,:,0] = 255
show(fig_cp)
```

```python
img = np.ones((512, 512, 3), dtype = np.uint8)
show(img)
```

# 感兴趣区域 ROI

在图像处理过程中，我们可能会对图像的某一个特定区域感兴趣，该区域被称为感兴趣区域（Region of Interest, ROI）。在设定感兴趣区域ROI后，就可以对该区域进行整体操作。

```python
fig = cv2.imread('images/highway.jpg', 1)
```

```python
show(fig)
```

```python
roi = fig[-400:-200, 200:400, :]
show(roi)
```

# 通道操作

## 通道拆分

```python
r, g, b = cv2.split(fig)
```

```python
show(r, 'R')
```

```python
show(g, 'G')
```

```python
show(b, 'B')
```

## 通道合并

```python
img = cv2.merge((r, g, b))
```

```python
show(img)
```

```python
img = cv2.merge((b, g, r))
show(img)
```
