# 导出

数据也可以被输出为分隔符格式的文本。

利用DataFrame的to_csv方法，我们可以将数据写到一个以逗号分隔的文件中。

```python
writer = pd.ExcelWriter('output.xlsx')
df.to_excel(writer)
```

```python
%cat output.xlsx
```

```python
df.to_csv('out.csv')
```
